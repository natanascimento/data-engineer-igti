## Média da nota em Ciências Humanas dos alunos do sexo MASCULINO?

```
%sql
SELECT AVG(NU_NOTA_CH) AS MEDIA_CIENCIASHUM FROM fEnem WHERE TP_SEXO = 'M';
```
